```python
num_str = input("Enter a number: ")
num = int(num_str)

if num % 2 == 0:
    print(f"{num} is an even number.")
else:
    print(f"{num} is an odd number.")
```

```python
print("Also to print number is whole number or not")
if (num % 5) == 0:
  print(f"{num} is a whole number.")
else:
  print(f"{num} is not a whole number.")
```
